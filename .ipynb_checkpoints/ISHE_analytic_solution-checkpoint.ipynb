```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.special import k0 
from scipy.integrate import quad
```

# checking the K0 function

```python
r = np.arange(0, 10, 0.01) # unit um
Ls = 5 # spin relaxation length L_s = \sqrt(D_n \tau_s)

plt.ylabel('K0(r/Ls)')
plt.xlabel('r (um)')
plt.plot(r,k0(r/1),'r',label = "Ls =1 um")
plt.plot(r,k0(r/3),'g',label = "Ls =3 um")
plt.plot(r,k0(r/5),'b',label = "Ls =5 um")
plt.legend()
```

```python
L = 10. # channel length
b = 10. # channel width
x0 = 1. # laser location by default
y0 = 1.
dy = 0.001
XMax = 200 # scan points in X
YMax = 200 # scan points in Y

I2D = np.zeros( (XMax,YMax) )

def s(x,y): #spin polarization density
    return k0(np.sqrt((x-x0)**2+(y-y0)**2)/Ls)

def ds_dyL(y):
    return (s(0,y+dy)-s(0,y))/dy

def ds_dyR(y):
    return (s(L,y+dy)-s(L,y))/dy
```

```python
#find photocurrent
for i in range(XMax):
    for j in range(YMax):
        x0 = L*(i + 0.5)/XMax # add 0.5 to center to the grid
        y0 = b*(j + 0.5)/YMax
        IL = quad(ds_dyL, 0, b)
        IR = quad(ds_dyR, 0, b)
        I2D[i,j] = IL[0]+IR[0]
    
# generate 2d grids for the x & y bounds
xi = np.arange(0, XMax) * L / XMax
yi = np.arange(0, YMax) * b / XMax
X1, Y1 = np.meshgrid(xi, yi)
    
#2D plot
fig1, ax1 = plt.subplots(figsize=(6, 5))
    
ax1.set_title("Helicity Dependent Current with Ls = " +str(Ls)+" um")
p = ax1.pcolor (X1, Y1, I2D, cmap ='RdBu')
ax1.set_ylabel('y (um)')
ax1.set_xlabel('x (um)')
cbar = fig1.colorbar(p)
cbar.set_label('Current')
```

# Circular Contact

```python

# helicity dependent photocurrent mapping for point contacts

L = 10. # channel length
b = 10. # channel width
x0 = 1. # laser location by default
y0 = 1.
dy = 0.001
XMax = 200 # scan points in X
YMax = 200 # scan points in Y

# parameters for disk contancts of radius R positioned at (xL, yL) and (xR, yR)
R = 1
xL = - R - 0.1 # Question: why do we need to -0.1 here?
yL = 0
xR = L + R + 0.1   # Question: why do we need to +0.1 here?
yR = 0

# differential steps
dx = 0.001
dy = 0.001

I2D = np.zeros( (XMax,YMax) )

def s(x,y): #spin polarization density
    return k0(np.sqrt((x-x0)**2+(y-y0)**2)/Ls)

def ds_dy(x, y):
    return (s(x, y + dy/2)-s(x, y - dy/2))/dy

def ds_dx(x, y):
    return (s(x + dx/2, y)-s(x - dx/2, y))/dx

def integrand(phi, xc, yc):
    # calculate x, y at the boundary of the disk
    x = xc + R*np.cos(phi)
    y = yc + R*np.sin(phi)
    return (ds_dy(x,y)*np.cos(phi) - ds_dx(x,y)*np.sin(phi))*R
```

```python
#find photocurrent
for i in range(XMax):
    for j in range(YMax):
        x0 = L*(i + 0.5)/XMax # add 0.5 to center to the grid
        y0 = -b + 2*b*(j + 0.5)/YMax
        IL = quad(integrand, 0, 2*np.pi, args=(xL,yL))
        IR = quad(integrand, 0, 2*np.pi, args=(xR,yR))
        I2D[i,j] = IL[0]+IR[0]
    
# generate 2d grids for the x & y bounds
xi = np.arange(0, XMax) * L / XMax
yi = np.arange(0, YMax) * 2 * b / YMax - b
X1, Y1 = np.meshgrid(xi, yi)
    
#2D plot
fig1, ax1 = plt.subplots(figsize=(6, 10))
    
ax1.set_title("Helicity Dependent Current with Ls = " +str(Ls)+" um")
p = ax1.pcolor (X1, Y1, I2D, cmap ='RdBu')
ax1.set_ylabel('y (um)')
ax1.set_xlabel('x (um)')
cbar = fig1.colorbar(p)
cbar.set_label('Current')
```
